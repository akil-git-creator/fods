```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('inventory_dataset.csv')
df
```

```python
df.isnull().any()
```

```python
null_columns = df.columns[df.isnull().any()]
```

```python
print("Columns with null values:", null_columns)
```

```python
median_purchase_price = df['Purchase_Price'].median()
median_purchase_price
```

```python
df['Purchase_Price'] = df['Purchase_Price'].fillna(median_purchase_price)
df
```

```python
mean_sale_price = df['Sale_Price'].mean()
mean_sale_price
```

```python
df['Sale_Price'] = df['Sale_Price'].fillna(mean_sale_price)
df
```

```python
df = df.dropna()
df
```

```python
Q1 = df['Sale_Price'].quantile(0.25)
Q3 = df['Sale_Price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df = df[~((df['Sale_Price'] < lower_bound) | (df['Sale_Price'] > upper_bound))]
df
```

```python
scaler = MinMaxScaler()
```

```python
df.loc[:, 'Quantity_In_Stock'] = scaler.fit_transform(df[['Quantity_In_Stock']])
```

```python
df.loc[:, 'Quantity_In_Stock'] = df['Quantity_In_Stock'].astype(int)
```

```python
df.loc[:, 'Discontinued'] = df['Discontinued'].astype(int)
```

```python
df.to_csv('cleaned_inventory.csv', index=False)
```
