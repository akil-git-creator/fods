```python
import pandas as pd
```

```python
dd=pd.read_csv('ecommerce_dataset (1).csv')
de=pd.read_csv('ecommerce_dataset.csv')
```

```python
merged_tab=pd.merge(de,dd,on='ProductID')
merged_tab.head(5)
```

```python
merged_tab.drop_duplicates()
merged_tab.head(5)
```

```python
q1=merged_tab['OrderPrice'].quantile(0.75)
q3=merged_tab['OrderPrice'].quantile(0.25)
```

```python
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
```

```python
outlier=(merged_tab[['OrderPrice']]<lb)|( merged_tab[['OrderPrice']]>ub)
outlier
```

```python
clean=(~merged_tab.isin(outlier))
clean
```

```python
sal_min=merged_tab['OrderPrice'].min()
sal_max=merged_tab['OrderPrice'].max()
```

```python
merged_tab['OrderPrice']=(merged_tab['OrderPrice']-sal_min)/(sal_max-sal_min)
merged_tab
```

```python
sale_min=merged_tab['OrderQuantity'].min()
sale_max=merged_tab['OrderQuantity'].max()
```

```python
merged_tab['OrderQuantity']=(merged_tab['OrderQuantity']-sal_min)/(sal_max-sal_min)
merged_tab
```

```python
dummies=pd.get_dummies(merged_tab,columns=['Category'])
dummies
```

```python
dummies=pd.get_dummies(merged_tab,columns=['Category'],drop_first=False)
dummies
```
